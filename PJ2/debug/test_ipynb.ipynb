```python
from HMM import HMM
from dataloader import dataloader
```

```python
dataloader = dataloader("NER/Chinese/train.txt","NER/Chinese/validation.txt")
```

```python
data, label=dataloader.get_data_list()
```

```python
state_num = dataloader.get_label_num()
obs_num = dataloader.get_data_num()
print(obs_num)
print(state_num)
```

```python
hmm = HMM(obs_num=obs_num, state_num=state_num)
```

```python
tran,emi,init=hmm.train(data, label)
```

```python
tran
```

```python

val_data, val_label = dataloader.get_test_data_list()
hmm.package_forward(val_data)
result=hmm.package_backward(val_data)
```

```python
len(result)
```

```python
dataloader.print_map()
```

```python
acc=[]
for i in range(len(result)):
    num_combined = [str(d)+" "+str(e) for d,e in zip(result[i], val_label[i])]
    #print("\nthe accuracy is: ",sum([1 for d,e in zip(result[i], val_label[i]) if int(d)==int(e)])/len(val_label[i]))
    acc.append(sum([1 for d,e in zip(result[i], val_label[i]) if int(d)==int(e)])/len(val_label[i]))
    if acc[-1]<0.5:
        print(num_combined)
        print(i)
        print("the accuracy is: ",sum([1 for d,e in zip(result[i], val_label[i]) if int(d)==int(e)])/len(val_label[i]))
print("the average accuracy is: ", sum(acc)/len(acc))
```
